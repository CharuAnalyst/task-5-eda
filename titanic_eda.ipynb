# 🚢 Titanic Dataset — Exploratory Data Analysis (EDA)
### Task 5 | Tools: Python, Pandas, Matplotlib, Seaborn
---
**Objective:** Extract insights using visual and statistical exploration of the Titanic passenger dataset.  
**Dataset:** 891 passengers | 10 features | Binary target: `survived` (0 = No, 1 = Yes)

## 1. 📦 Import Libraries & Load Dataset

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Style settings
plt.rcParams['figure.dpi'] = 120
plt.rcParams['font.family'] = 'sans-serif'
sns.set_theme(style='whitegrid', palette='muted')

print("Libraries loaded successfully ✅")
```

```python
# Load the Titanic dataset
# Option 1: Load from seaborn (requires internet)
# df = sns.load_dataset('titanic')

# Option 2: Load from CSV (downloaded/local)
df = pd.read_csv('titanic.csv')

print(f"Dataset Shape: {df.shape}")
print(f"\nColumns: {list(df.columns)}")
df.head()
```

## 2. 📊 Dataset Overview — .info() and .describe()

```python
# Basic info about data types and missing values
print("=" * 55)
print("DATASET INFO (.info())")
print("=" * 55)
df.info()
```

```python
# Statistical summary of numerical features
print("=" * 55)
print("STATISTICAL SUMMARY (.describe())")
print("=" * 55)
df.describe().round(2)
```

```python
# Check missing values
missing = df.isnull().sum()
missing_pct = (df.isnull().sum() / len(df) * 100).round(1)
missing_df = pd.DataFrame({'Missing Count': missing, 'Missing %': missing_pct})
missing_df = missing_df[missing_df['Missing Count'] > 0]
print("Missing Values:")
print(missing_df)
```

```python
# Value counts for categorical columns
print("Survived:\n", df['survived'].value_counts(), "\n")
print("Sex:\n", df['sex'].value_counts(), "\n")
print("Pclass:\n", df['pclass'].value_counts().sort_index(), "\n")
print("Embarked:\n", df['embarked'].value_counts())
```

## 3. 🎯 Survival Analysis

```python
# Figure 1: Overall Survival Count
fig, ax = plt.subplots(figsize=(6, 4))
colors_surv = ['#e74c3c', '#2ecc71']
counts = df['survived'].value_counts().sort_index()
bars = ax.bar(['Did Not Survive', 'Survived'], counts.values,
              color=colors_surv, edgecolor='white', linewidth=1.5, width=0.5)
for bar, v in zip(bars, counts.values):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 5,
            f'{v}\n({v/len(df)*100:.1f}%)', ha='center', fontsize=11, fontweight='bold')
ax.set_title('Overall Survival Distribution', fontsize=14, fontweight='bold')
ax.set_ylabel('Number of Passengers')
ax.set_ylim(0, counts.max() * 1.25)
ax.grid(axis='y', alpha=0.3)
ax.spines['top'].set_visible(False); ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.show()

# Observation
print("\n📌 OBSERVATION:")
print(f"  • {counts[0]} passengers ({counts[0]/len(df)*100:.1f}%) did NOT survive")
print(f"  • {counts[1]} passengers ({counts[1]/len(df)*100:.1f}%) SURVIVED")
print("  • Dataset is imbalanced — important to note for any future ML modeling")
```

```python
# Figure 2: Survival by Sex
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Bar chart
sex_surv = df.groupby(['sex', 'survived']).size().unstack()
sex_surv.plot(kind='bar', ax=axes[0], color=['#e74c3c', '#2ecc71'],
              edgecolor='white', width=0.6)
axes[0].set_title('Survival Count by Sex', fontsize=13, fontweight='bold')
axes[0].set_xticklabels(['Female', 'Male'], rotation=0)
axes[0].set_xlabel('Sex'); axes[0].set_ylabel('Count')
axes[0].legend(['Did Not Survive', 'Survived'], frameon=False)
axes[0].grid(axis='y', alpha=0.3)
axes[0].spines['top'].set_visible(False); axes[0].spines['right'].set_visible(False)

# Survival rate
sex_rate = df.groupby('sex')['survived'].mean() * 100
axes[1].bar(sex_rate.index, sex_rate.values, color=['#9b59b6', '#3498db'], width=0.4)
for i, (sex, rate) in enumerate(sex_rate.items()):
    axes[1].text(i, rate + 1, f'{rate:.1f}%', ha='center', fontsize=12, fontweight='bold')
axes[1].set_title('Survival Rate by Sex', fontsize=13, fontweight='bold')
axes[1].set_ylabel('Survival Rate (%)'); axes[1].set_ylim(0, 85)
axes[1].set_xticklabels(['Female', 'Male'])
axes[1].grid(axis='y', alpha=0.3)
axes[1].spines['top'].set_visible(False); axes[1].spines['right'].set_visible(False)

plt.tight_layout()
plt.show()

print("\n📌 OBSERVATION:")
for sex in sex_rate.index:
    print(f"  • {sex.capitalize()} survival rate: {sex_rate[sex]:.1f}%")
print("  • Women were ~3.5x more likely to survive than men")
print("  • Strong evidence of 'women and children first' evacuation policy")
```

```python
# Figure 3: Survival by Passenger Class
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

pc_surv = df.groupby(['pclass', 'survived']).size().unstack()
pc_surv.plot(kind='bar', ax=axes[0], color=['#e74c3c', '#2ecc71'],
             edgecolor='white', width=0.6)
axes[0].set_title('Survival Count by Passenger Class', fontsize=13, fontweight='bold')
axes[0].set_xticklabels(['1st Class', '2nd Class', '3rd Class'], rotation=0)
axes[0].set_xlabel('Passenger Class'); axes[0].set_ylabel('Count')
axes[0].legend(['Did Not Survive', 'Survived'], frameon=False)
axes[0].grid(axis='y', alpha=0.3)
axes[0].spines['top'].set_visible(False); axes[0].spines['right'].set_visible(False)

pc_rate = df.groupby('pclass')['survived'].mean() * 100
axes[1].bar(['1st', '2nd', '3rd'], pc_rate.values,
            color=['#f1c40f', '#e67e22', '#e74c3c'], width=0.4)
for i, rate in enumerate(pc_rate.values):
    axes[1].text(i, rate + 1, f'{rate:.1f}%', ha='center', fontsize=12, fontweight='bold')
axes[1].set_title('Survival Rate by Passenger Class', fontsize=13, fontweight='bold')
axes[1].set_ylabel('Survival Rate (%)'); axes[1].set_ylim(0, 80)
axes[1].grid(axis='y', alpha=0.3)
axes[1].spines['top'].set_visible(False); axes[1].spines['right'].set_visible(False)

plt.tight_layout()
plt.show()

print("\n📌 OBSERVATION:")
class_names = {1: '1st Class', 2: '2nd Class', 3: '3rd Class'}
for pc in sorted(pc_rate.index):
    print(f"  • {class_names[pc]}: {pc_rate[pc]:.1f}% survival rate")
print("  • Clear socioeconomic gradient — wealthy passengers had higher survival")
```

## 4. 📈 Distributions — Histograms & Boxplots

```python
# Figure 4: Age Distribution by Survival
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Overlapping histogram
for s, color, label in [(0, '#e74c3c', 'Did Not Survive'), (1, '#2ecc71', 'Survived')]:
    subset = df[df['survived'] == s]['age'].dropna()
    axes[0].hist(subset, bins=30, alpha=0.6, color=color, label=label, edgecolor='white')
axes[0].axvline(df['age'].median(), color='steelblue', linestyle='--', linewidth=2,
                label=f'Median: {df["age"].median():.1f} yrs')
axes[0].set_title('Age Distribution by Survival', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Age'); axes[0].set_ylabel('Count')
axes[0].legend(frameon=False); axes[0].grid(alpha=0.3)
axes[0].spines['top'].set_visible(False); axes[0].spines['right'].set_visible(False)

# Boxplot by survived
df.boxplot(column='age', by='survived', ax=axes[1],
           boxprops=dict(color='steelblue'), medianprops=dict(color='red', linewidth=2),
           flierprops=dict(marker='o', alpha=0.3))
axes[1].set_title('Age Boxplot by Survival', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Survived (0=No, 1=Yes)'); axes[1].set_ylabel('Age')
plt.suptitle('')
plt.tight_layout()
plt.show()

print("\n📌 OBSERVATION:")
for s, label in [(0, 'Non-survivors'), (1, 'Survivors')]:
    ages = df[df['survived'] == s]['age'].dropna()
    print(f"  • {label}: mean={ages.mean():.1f}, median={ages.median():.1f}, std={ages.std():.1f}")
print(f"  • {df['age'].isnull().sum()} missing age values ({df['age'].isnull().mean()*100:.1f}%)")
```

```python
# Figure 5: Fare Distributions — Boxplots & Histograms
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Boxplot by pclass
data_box = [df[df['pclass'] == i]['fare'].dropna() for i in [1, 2, 3]]
bp = axes[0].boxplot(data_box, patch_artist=True, notch=False,
                     medianprops=dict(color='white', linewidth=2), widths=0.5)
for patch, color in zip(bp['boxes'], ['#3498db', '#f39c12', '#e74c3c']):
    patch.set_facecolor(color); patch.set_alpha(0.75)
axes[0].set_xticklabels(['1st Class', '2nd Class', '3rd Class'])
axes[0].set_title('Fare Distribution by Class (Boxplot)', fontsize=13, fontweight='bold')
axes[0].set_ylabel('Fare (£)'); axes[0].grid(axis='y', alpha=0.3)
axes[0].spines['top'].set_visible(False); axes[0].spines['right'].set_visible(False)

# Histogram (log-scale for fare)
axes[1].hist(df['fare'].clip(upper=200), bins=40, color='#3498db', edgecolor='white', alpha=0.8)
axes[1].axvline(df['fare'].median(), color='red', linestyle='--', linewidth=2,
                label=f'Median: £{df["fare"].median():.1f}')
axes[1].set_title('Overall Fare Distribution', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Fare (£, capped at 200)'); axes[1].set_ylabel('Count')
axes[1].legend(frameon=False); axes[1].grid(alpha=0.3)
axes[1].spines['top'].set_visible(False); axes[1].spines['right'].set_visible(False)

plt.tight_layout()
plt.show()

print("\n📌 OBSERVATION:")
for pc, label in [(1,'1st'),(2,'2nd'),(3,'3rd')]:
    fares = df[df['pclass']==pc]['fare']
    print(f"  • {label} class: median=£{fares.median():.1f}, max=£{fares.max():.1f}")
print("  • Extreme right skew in fare distribution — many outliers in 1st class")
```

## 5. 🔥 Correlation Heatmap

```python
# Figure 6: Correlation Heatmap
fig, ax = plt.subplots(figsize=(8, 6))
num_cols = ['survived', 'pclass', 'age', 'sibsp', 'parch', 'fare']
corr = df[num_cols].corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, fmt='.2f', cmap='RdYlGn',
            center=0, linewidths=0.5, ax=ax,
            annot_kws={'size': 11}, vmin=-1, vmax=1)
ax.set_title('Correlation Heatmap — Numeric Features', fontsize=14, fontweight='bold', pad=12)
plt.tight_layout()
plt.show()

print("\n📌 OBSERVATION:")
surv_corr = corr['survived'].drop('survived').sort_values(key=abs, ascending=False)
print("  Correlations with Survived:")
for feat, val in surv_corr.items():
    direction = '↑ positive' if val > 0 else '↓ negative'
    print(f"  • {feat}: {val:.2f} ({direction})")
print("  • Pclass has strongest negative correlation: higher class number → lower survival")
print("  • Fare has moderate positive correlation: paying more → higher survival chances")
```

## 6. 🔵 Scatter Plots & Pair Analysis

```python
# Figure 7: Fare vs Age Scatterplot
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

for s, color, label, marker in [(0, '#e74c3c', 'Did Not Survive', 'x'),
                                  (1, '#2ecc71', 'Survived', 'o')]:
    subset = df[df['survived'] == s].dropna(subset=['age', 'fare'])
    axes[0].scatter(subset['age'], subset['fare'], c=color, alpha=0.45,
                    s=25, marker=marker, label=label)
axes[0].set_title('Fare vs Age by Survival', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Age'); axes[0].set_ylabel('Fare (£)')
axes[0].set_ylim(-5, 250); axes[0].legend(frameon=False); axes[0].grid(alpha=0.2)
axes[0].spines['top'].set_visible(False); axes[0].spines['right'].set_visible(False)

# Fare vs Age, colored by pclass
colors_pc = {1: '#3498db', 2: '#f39c12', 3: '#e74c3c'}
for pc in [1, 2, 3]:
    s = df[df['pclass'] == pc].dropna(subset=['age', 'fare'])
    axes[1].scatter(s['age'], s['fare'], c=colors_pc[pc], alpha=0.45,
                    s=25, label=f'Class {pc}')
axes[1].set_title('Fare vs Age by Passenger Class', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Age'); axes[1].set_ylabel('Fare (£)')
axes[1].set_ylim(-5, 250); axes[1].legend(frameon=False); axes[1].grid(alpha=0.2)
axes[1].spines['top'].set_visible(False); axes[1].spines['right'].set_visible(False)

plt.tight_layout()
plt.show()

print("\n📌 OBSERVATION:")
print("  • Survivors cluster at higher fare values across all ages")
print("  • Non-survivors dominate the low-fare (3rd class) band")
print("  • No strong linear age-fare relationship; pattern is driven by class")
```

```python
# Seaborn Pairplot (subset of features)
pair_cols = ['survived', 'pclass', 'age', 'fare']
pair_df = df[pair_cols].dropna()
pair_df['Survived'] = pair_df['survived'].map({0: 'No', 1: 'Yes'})

g = sns.pairplot(pair_df, hue='Survived', palette={'No': '#e74c3c', 'Yes': '#2ecc71'},
                 vars=['pclass', 'age', 'fare'], plot_kws={'alpha': 0.4, 's': 20},
                 diag_kind='hist')
g.fig.suptitle('Pairplot: Pclass, Age & Fare by Survival', y=1.02, fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print("\n📌 OBSERVATION:")
print("  • Pairplot confirms pclass and fare separation between survivors/non-survivors")
print("  • Age shows moderate separation (young children in survived group)")
print("  • Fare and pclass are negatively correlated (higher class = lower number = higher fare)")
```

## 7. 🌍 Embarkation & Family Analysis

```python
# Embarkation & Family features
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Embarkation survival rate
emb_rate = df.groupby('embarked')['survived'].mean() * 100
emb_labels = {'S': 'Southampton', 'C': 'Cherbourg', 'Q': 'Queenstown'}
emb_rate.index = [emb_labels.get(e, e) for e in emb_rate.index]
emb_rate.sort_values(ascending=False).plot(kind='bar', ax=axes[0],
    color=['#3498db', '#9b59b6', '#e67e22'], edgecolor='white', width=0.5)
for i, (_, v) in enumerate(emb_rate.sort_values(ascending=False).items()):
    axes[0].text(i, v+0.5, f'{v:.1f}%', ha='center', fontsize=10, fontweight='bold')
axes[0].set_title('Survival Rate by Embarkation', fontsize=12, fontweight='bold')
axes[0].set_ylabel('Survival Rate (%)'); axes[0].set_ylim(0, 65)
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=15)
axes[0].grid(axis='y', alpha=0.3)
axes[0].spines['top'].set_visible(False); axes[0].spines['right'].set_visible(False)

# SibSp survival rate
sibsp_rate = df.groupby('sibsp')['survived'].mean() * 100
sibsp_rate.plot(kind='bar', ax=axes[1], color='#3498db', edgecolor='white', width=0.6)
axes[1].set_title('Survival Rate by SibSp', fontsize=12, fontweight='bold')
axes[1].set_xlabel('# Siblings/Spouses'); axes[1].set_ylabel('Survival Rate (%)')
axes[1].set_xticklabels(sibsp_rate.index, rotation=0)
axes[1].grid(axis='y', alpha=0.3)
axes[1].spines['top'].set_visible(False); axes[1].spines['right'].set_visible(False)

# Alone vs not
alone_rate = df.groupby('alone')['survived'].mean() * 100
axes[2].bar(['Traveling with Family', 'Traveling Alone'], alone_rate.values,
            color=['#2ecc71', '#e74c3c'], width=0.4)
for i, v in enumerate(alone_rate.values):
    axes[2].text(i, v+0.5, f'{v:.1f}%', ha='center', fontsize=11, fontweight='bold')
axes[2].set_title('Survival Rate: Alone vs With Family', fontsize=12, fontweight='bold')
axes[2].set_ylabel('Survival Rate (%)'); axes[2].set_ylim(0, 55)
axes[2].grid(axis='y', alpha=0.3)
axes[2].spines['top'].set_visible(False); axes[2].spines['right'].set_visible(False)

plt.tight_layout()
plt.show()

print("\n📌 OBSERVATIONS:")
print("  • Cherbourg passengers had highest survival — mostly 1st-class travelers")
print("  • Small families (1-2 siblings/spouses) had better survival than solo travelers or large families")
print("  • Traveling alone was associated with lower survival rates")
```

## 8. 📋 Summary of Findings

| # | Finding | Key Insight |
|---|---------|-------------|
| 1 | **Survival Rate** | Only 35.9% survived — dataset is imbalanced |
| 2 | **Gender** | Women: 67% survival vs Men: 19% — strongest predictor |
| 3 | **Passenger Class** | 1st: 57.5% > 2nd: 45.2% > 3rd: 23.9% survival |
| 4 | **Age** | Children had better survival odds; median age ~29.8 years |
| 5 | **Fare** | Higher fare → higher survival (proxy for wealth/class) |
| 6 | **Embarkation** | Cherbourg highest survival due to 1st-class concentration |
| 7 | **Family Size** | Small families fared better; solo travelers at disadvantage |
| 8 | **Missing Data** | 19% of Age values missing — needs imputation before ML |
| 9 | **Correlations** | Pclass (-0.33) & Fare (+0.26) most correlated with survival |
| 10 | **Anomalies** | Extreme fare outliers (up to £299); Some passengers 0.5 yrs old |

### 🎯 Conclusion
> The Titanic disaster was not a random tragedy — survival was strongly shaped by **gender, socioeconomic class, and age**.  
> These three features should form the foundation of any predictive model.  
> Data quality issues (missing age) must be resolved via imputation before ML training.
